# **Model**


---



 **Base Model**

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
lin_model= LinearRegression()
lin_model.fit(X_train,y_train)
y_pred_lin=lin_model.predict(X_test)
```

```python
rf_model=RandomForestRegressor()
rf_model.fit(X_train,y_train)
y_pred_rf=rf_model.predict(X_test)
```

```python
def regression_mertices(y_true,y_pred):
 mae =mean_absolute_error(y_true,y_pred)
 mse=mean_squared_error(y_true,y_pred)
 rmse=np.sqrt(mse)
 r2=r2_score(y_true,y_pred)
 return mae,rmse,r2
```

```python
mae_lin,rmse_lin,r2_lin=regression_mertices(y_test,y_pred_lin)
mae_rf,rmse_rf,r2_rf=regression_mertices(y_test,y_pred_rf)
print(f"Linear Regression Evaluation: MAE:{mae_lin}, RMSE:{rmse_lin}, r2_score:{r2_lin}")
print(f"Random Forest Regression Evaluation: MAE:{mae_rf}, RMSE:{rmse_rf}, r2_score:{r2_rf}")
```

**Observation on Base model**



1.  MAE (Mean Absolute Error):

   *   Linear Regression: 0.623
   *   Random Forest: 0.352

**Random Forest is, on average, only off by about 0.35 points — which is excellent.**


2.   RMSE (Root Mean Squared Error):

   *  Linear Regression: 0.782
   *   Random Forest: 0.536

**Random Forest again performs better — it makes fewer and smaller mistakes overall.**


3.  R² Score (Coefficient of Determination):

   * Linear Regression: 0.757
   * Random Forest: .891

**Random Forest explains nearly 89% of the variation in addiction level — that’s very strong.**

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

params= {
    'n_estimators': randint(100, 500),
    'max_depth': [10,20,30,40,50],
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 10),
    'max_features': ['sqrt', 'log2']
}

rf = RandomForestRegressor(random_state=42, n_jobs=-1)
random_search = RandomizedSearchCV(estimator=rf,param_distributions=params,n_iter=50,cv=5,scoring='r2', verbose=2,random_state=42,n_jobs=-1)
random_search.fit(X_train,y_train)

```

```python
print(f"Best Parameters: {random_search.best_params_}")
print(f"Best R2 Score: {random_search.best_score_}")
```

**Observation**

After applying hyperparameter optimization using RandomizedSearchCV, the model’s R² slightly decreased from 0.89 to 0.81.
This is expected because cross-validation provides a more robust and conservative estimate of model performance.
The tuned model may generalize better to unseen data, even if its internal validation score appears lower.

```python
from sklearn.model_selection import GridSearchCV
params_grid={
    'n_estimators':[250,300,320,350,400],
    'max_depth':[25,30,35],
    'min_samples_split':[2,3,4],
    'min_samples_leaf':[1,2,3],
    'max_features': ['sqrt', 'log2']
}
rf=RandomForestRegressor(random_state=42,n_jobs=-1)
grid_search=GridSearchCV(estimator=rf,param_grid=params_grid,cv=5,scoring='r2',n_jobs=-1,verbose=2)
grid_search.fit(X_train,y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best R2 Score:", grid_search.best_score_)
```

**OBSERVATION**

A fine-tuning step using GridSearchCV was applied around the best parameters found in RandomizedSearchCV.

The final model achieved an R² of 0.824, which is not the best progress yet, going forward to XGBOOST model

```python
from xgboost import XGBRegressor


xgb = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
```

```python
xgb.fit(X_train, y_train)

y_pred_xgb = xgb.predict(X_test)

mae = mean_absolute_error(y_test, y_pred_xgb)
r2 = r2_score(y_test, y_pred_xgb)
print(f"XGBoost MAE: {mae:.3f}, R²: {r2:.3f}")
```

```python
from scipy.stats import uniform
param_dist = {
    'n_estimators': randint(100, 500),
    'learning_rate': uniform(0.01, 0.2),
    'max_depth': randint(3, 10),
    'subsample': uniform(0.6, 0.4),
    'colsample_bytree': uniform(0.6, 0.4),
    'min_child_weight': randint(1, 10)
}

xgb=XGBRegressor(random_state=42, n_jobs=-1)
random_search_xgb=RandomizedSearchCV(estimator=xgb, param_distributions=param_dist, cv=5,scoring='r2',random_state=42,n_jobs=-1,verbose=2)
random_search_xgb.fit(X_train,y_train)
print("Best Parameters:", random_search_xgb.best_params_)

```

```python
best_xgb = random_search_xgb.best_estimator_
y_pred_xgb = best_xgb.predict(X_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred_xgb)
r2 = r2_score(y_test, y_pred_xgb)

print(f"XGBoost Regression Evaluation: MAE:{mae}, R²:{r2}")
```

```python
best_params = random_search_xgb.best_params_

param_dist2 = {
    'n_estimators': [best_params['n_estimators'] - 30,
                     best_params['n_estimators'],
                     best_params['n_estimators'] + 30],
    'learning_rate': [best_params['learning_rate'] * 0.8,
                      best_params['learning_rate'],
                      best_params['learning_rate'] * 1.2],
    'max_depth': [best_params['max_depth'] - 1,
                  best_params['max_depth'],
                  best_params['max_depth'] + 1],
    'subsample': [max(0.6, best_params['subsample'] - 0.1),
                  best_params['subsample'],
                  min(1.0, best_params['subsample'] + 0.1)],
    'colsample_bytree': [max(0.6, best_params['colsample_bytree'] - 0.1),
                         best_params['colsample_bytree'],
                         min(1.0, best_params['colsample_bytree'] + 0.1)]

}
grid_search_xgb = GridSearchCV(estimator=xgb, param_grid=param_dist2, cv=5, scoring='r2', n_jobs=-1, verbose=2)
grid_search_xgb.fit(X_train, y_train)

best_xgb_grid = grid_search_xgb.best_estimator_
y_pred_xgb_grid = best_xgb_grid.predict(X_test)

mae = mean_absolute_error(y_test, y_pred_xgb_grid)
r2 = r2_score(y_test, y_pred_xgb_grid)

print("Best Parameters from Grid Search:", grid_search_xgb.best_params_)
print(f"XGBoost last tuned: MAE:{mae}, R²:{r2}")
```

```python
if hasattr(best_xgb, 'feature_importances_'):
    importances = best_xgb.feature_importances_
    feature_names = X_train.columns
    imp_df = pd.DataFrame({'feature': feature_names, 'importance': importances})
    imp_df = imp_df.sort_values('importance', ascending=False).head(20)

    plt.figure(figsize=(8,6))
    sns.barplot(data=imp_df, x='importance', y='feature')
    plt.title("Top 20 Feature Importances — XGBoost")
    plt.xlabel("Importance")
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Ensure regression_mertices is defined or imported if it's in a separate cell
# (Assuming it's defined in the notebook already)

# Recalculate XGBoost metrics using y_pred_xgb_grid from the last GridSearchCV
mae_xgb_final, rmse_xgb_final, r2_xgb_final = regression_mertices(y_test, y_pred_xgb_grid)

print(f"Linear Regression Evaluation: MAE:{mae_lin}, RMSE:{rmse_lin}, r2_score:{r2_lin}")
print(f"Random Forest Regression Evaluation: MAE:{mae_rf}, RMSE:{rmse_rf}, r2_score:{r2_rf}")
print(f"XGBoost Regression Evaluation (tuned): MAE:{mae_xgb_final}, RMSE:{rmse_xgb_final}, r2_score:{r2_xgb_final}")

# -----------------------------
# 2) PCA TRANSFORMATION (2D)
# -----------------------------
pca = PCA(n_components=2)
# Fit PCA on the training data and transform the test data
pca.fit(X_train) # Fit on training data to prevent data leakage
X_test_pca = pca.transform(X_test) # Transform test data
explained_var = pca.explained_variance_ratio_

# -----------------------------
# 3) CREATE DASHBOARD FIGURE
# -----------------------------
fig = plt.figure(figsize=(14, 6))

# ---- PCA Plot ----
ax1 = fig.add_subplot(1, 2, 1)
# Plot X_test_pca components, colored by y_test
scatter = ax1.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_test, alpha=0.7, cmap='viridis') # Added cmap for better visualization
ax1.set_title("PCA Visualization (2 Components of Test Set)")
ax1.set_xlabel(f"PC1 ({explained_var[0]*100:.2f}% var)")
ax1.set_ylabel(f"PC2 ({explained_var[1]*100:.2f}% var)")
cbar = plt.colorbar(scatter, ax=ax1)
cbar.set_label("Target Value")

# ---- Metrics Comparison ----
ax2 = fig.add_subplot(1, 2, 2)

metrics_labels = ["MAE", "RMSE", "R2"] # Define metrics_labels before using it
x = np.arange(len(metrics_labels))
width = 0.25 # Adjust width to accommodate 3 bars

linear_metrics = [mae_lin, rmse_lin, r2_lin]
rf_metrics = [mae_rf, rmse_rf, r2_rf]
xgb_metrics = [mae_xgb_final, rmse_xgb_final, r2_xgb_final] # Using final tuned XGBoost metrics

ax2.bar(x - width, linear_metrics, width, label="Linear Regression")
ax2.bar(x, rf_metrics, width, label="Random Forest")
ax2.bar(x + width, xgb_metrics, width, label="XGBoost")

ax2.set_title("Model Performance Metrics")
ax2.set_xticks(x, metrics_labels)
ax2.set_ylabel("Metric Value")
ax2.grid(axis="y", linestyle="--", alpha=0.3)
ax2.legend()

# -----------------------------
# Final render
# -----------------------------
plt.tight_layout()
plt.show()
```
